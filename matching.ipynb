```python
import pandas as pd
import tabula
import numpy as np
```

```python
dfl = tabula.read_pdf("matching.pdf", lattice=True, pages='all')
dfl
```

```python
hospList = ['株式会社日立製作所日立総合病院', '長野赤十字病院', '島根県立中央病院',
            '益田赤十字病院', '大田市立病院', '岡山赤十字病院', '松山赤十字病院',
            '大分赤十字病院', '佐世保共済病院', '会津中央病院', '市川総合病院',
            '竹田総合病院','九州大学病院']

ndf = pd.DataFrame(columns=['県名' '病院\r番号' '病院名称' 'プログラム\r番号' 'プログラム名称' '定\r員' 'マ\rッ\rチ\r者\r数'
 '空\r席\r数', 'allmatch'])

def judgeF(hospitalName):
    if hospitalName in hospList:
        return True
    return False

def judgeL(left):
    if left > 0.0:
        return True
    else:
        return False

for i in range(len(dfl)):
    df = dfl[i]
    df['allmatch'] = df['空\r席\r数'].apply(judgeL)
    df = df[df['allmatch']==True]
    ndf = pd.concat([ndf, df])

ndf
ndf.to_excel('left.xlsx')
```
